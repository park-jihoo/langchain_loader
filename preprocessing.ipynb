### PDF Files

```python
!uv pip install -qU pyPdf
```

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("document/pdf_sample.pdf")
pdf_docs = loader.load()

print(f"{pdf_docs[0].metadata=}\n")
print(pdf_docs[0].page_content)
```

### Web Pages

```python
!uv pip install beautifulsoup4
```

```python
from langchain_community.document_loaders import WebBaseLoader
import bs4

loader = WebBaseLoader(
    web_paths=["https://www.lgcns.com"],
    bs_kwargs={"parse_only": bs4.SoupStrainer("body")},
    bs_get_text_kwargs={"separator": "\n", "strip": True}
)

web_docs = loader.load()

print("\n")
print(f"{web_docs[0].metadata=}\n")
print(web_docs[0].page_content)
```

### CSV Data

```python
from langchain_community.document_loaders import CSVLoader

csv_loader = CSVLoader(
    file_path="document/csv_sample.csv",
    csv_args={"delimiter": ","},
    source_column="Player",
)


csv_docs = csv_loader.load()

for record in csv_docs[:2]:
    print(record)
```

### JSON Data

```python
!uv pip install jq
```

```python
from langchain_community.document_loaders import JSONLoader

def metadata_func(record: dict, metadata: dict) -> dict:
    metadata["Player"] = record["Player"]
    return metadata

json_loader = JSONLoader(
    file_path="document/json_sample.json",
    json_lines=True,
    content_key="Annual",
    jq_schema=".",
    metadata_func=metadata_func,
    text_content=False
)

json_docs = json_loader.load()

print(json_docs)
```
